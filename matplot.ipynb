### Problem statement 2: Analysis the behaviour of glucose and Bloodpressure for diabetic and non diabetic person

```python
import pandas as pd

dia=pd.read_csv('diabetes.csv')
diabetes=dia[dia.Outcome==0]
nonDiabetes=dia[dia.Outcome==1]

plt.figure(figsize=(20,10))

# diabatic
plt.subplot(1,2,1)
plt.scatter(diabetes.BloodPressure,diabetes.Glucose,c='r')
plt.xlabel('Blood Pressure')
plt.ylabel('Glucose')
plt.title('diabatic')

# non diabatic
plt.subplot(1,2,2)
plt.scatter(nonDiabetes.BloodPressure,nonDiabetes.Glucose,c='b')
plt.xlabel('Blood presure')
plt.ylabel('Glucose')
plt.title('non diabatic')
```
